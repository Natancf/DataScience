#Dados

```python
from random import randrange
```

```python
faturamento_empresa = []

for faturamento in range(12):
    faturamento_empresa.append(randrange(50, 900, 50))
faturamento_empresa
```

#Gráfico

```python
import matplotlib.pyplot as plt
```

```python
x = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
y = faturamento_empresa
plt.plot(x, y, marker='o')
plt.title('Faturamento da empresa em milhões')
plt.show()
```
